```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_excel (r'COMP5122M_Aug_2020.xlsx', sheet_name='Data')
```

```python
def Avg(lst): 
    return sum(lst) / len(lst)
```

```python
age = {}
leave = {}
remain = {}
income = {}
referendum = {}

for index, row in df.iterrows():
    referendum[row["ONS code"]] = row["Brexit referendum"]
    leave[row["ONS code"]] = row["Leave petition"]
    remain[row["ONS code"]] = row["Remain petition"]
    age[row["ONS code"]] = row["Aged 65+"]
    income[row["ONS code"]] = row["Median monthly earnings"]
sorted_dict = {k: v for k, v in sorted(referendum.items(), key=lambda item: item[1])}

c = list()
r = list()
l = list()
b = list()
a = list()
i = list()
for key in sorted_dict:
    c.append(key)
    r.append(remain[key])
    l.append(leave[key])
    b.append(referendum[key])
    a.append(age[key])
    i.append(income[key])
```

```python
age = {}
leave = {}
remain = {}
country = list()

for index, row in df.iterrows():
    if row["Country"] not in country:
        country.append(row["Country"])
    if row["Country"] not in leave:
        leave[row["Country"]] = row["Leave petition"]
    else:
        leave[row["Country"]] += row["Leave petition"]
    if row["Country"] not in remain:
        remain[row["Country"]] = row["Remain petition"]
    else:
        remain[row["Country"]] += row["Remain petition"]
    if row["Country"] not in age:
        age[row["Country"]] = row["Aged 65+"]
    else:
        age[row["Country"]] += row["Aged 65+"]
```
